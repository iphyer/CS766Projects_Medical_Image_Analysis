```python
import csv
import os
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np
```

```python
resultlist_all = []
with open('result1.csv', 'r') as file: 
    reader = csv.reader(file)
    resultlist_all.append(list(reader))

with open('result2.csv', 'r') as file: 
    reader = csv.reader(file)
    resultlist_all.append(list(reader))
    
with open('result3.csv', 'r') as file: 
    reader = csv.reader(file)
    resultlist_all.append(list(reader))
    
with open('result4.csv', 'r') as file: 
    reader = csv.reader(file)
    resultlist_all.append(list(reader))
    
with open('result5.csv', 'r') as file: 
    reader = csv.reader(file)
    resultlist_all.append(list(reader))
    
    
with open('testlabels.csv', 'r') as file: 
    reader = csv.reader(file)
    labellist = list(reader)

path = 'imgs'
files = os.listdir(path)
for i in range(1, len(files)):
    
    lab = labellist[i]
   
    #print(lab)
    file = files[i];
    print(file)
    im = np.array(Image.open(path + '/' + file), dtype=np.uint8)
    fig,ax = plt.subplots(1)

    ax.imshow(im)
    for j in range(5):
        res = resultlist_all[j][i-1]
        print(res)
        ymin = float(res[1])
        xmin = float(res[2])
        ymax = float(res[3])
        xmax = float(res[4])
        rect = patches.Rectangle((ymin, xmin), ymax-ymin, xmax-xmin, linewidth=3, edgecolor='r', facecolor='none')
        ax.add_patch(rect)
    ymi = int(lab[3])
    xmi = int(lab[2])
    yma = int(lab[5])
    xma = int(lab[4])
    
    
    rect2 = patches.Rectangle((ymi, xmi), yma-ymi, xma-xmi, linewidth=2, edgecolor='b', facecolor='none')
    
    ax.add_patch(rect2)
    plt.show()
```
