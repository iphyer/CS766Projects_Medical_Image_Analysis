```python
import csv
import os
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np
```

```python
with open('result.csv', 'r') as file: 
    reader = csv.reader(file)
    resultlist = list(reader)


with open('label.csv', 'r') as file: 
    reader = csv.reader(file)
    labellist = list(reader)

path = 'imgs'
files = os.listdir(path)
for i in range(len(files)):
    res = resultlist[i]
    lab = labellist[i]
    #print() 
    print(res)
    print(lab)
    file = files[i];
    print(file)
    im = np.array(Image.open(path + '/' + file), dtype=np.uint8)
    fig,ax = plt.subplots(1)

    ax.imshow(im)
    ymin = float(res[1])
    xmin = float(res[2])
    ymax = float(res[3])
    xmax = float(res[4])
    
    ymi = int(lab[1])
    xmi = int(lab[2])
    yma = int(lab[3])
    xma = int(lab[4])
    
    rect1 = patches.Rectangle((ymin, xmin), ymax-ymin, xmax-xmin, linewidth=3, edgecolor='r', facecolor='none')
    rect2 = patches.Rectangle((ymi, xmi), yma-ymi, xma-xmi, linewidth=2, edgecolor='b', facecolor='none')
    ax.add_patch(rect1)
    ax.add_patch(rect2)
    plt.show()
```
